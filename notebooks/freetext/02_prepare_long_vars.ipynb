```python
def get_var_file(varno:int, lfdn=False):
    basedir = '../../data/freetext'
    long_vars = sorted(
        [x for x in os.listdir(basedir) if 'long' in x],
          key=lambda x:int(x.split('_')[1]))
    this_file = [x for x in long_vars if int(x.split('_')[1]) == varno][0]
    print(this_file)
    if not lfdn:
        return pd.read_csv(f'{basedir}/{this_file}')[f'v_{varno}'].values
    else:
        return pd.read_csv(f'{basedir}/{this_file}').values
```

**Note:** This is context-dependent - the answers cannot be understood without knowing respondents' `TOP Problem` selections (whoever designed this survey this way: screw you).

```python
def prepare_with_dependencies(df, longvar, depvar):
    var = df[['lfdn', f'v_{depvar}', f'v_{longvar}']
      ][~df[f'v_{longvar}'].isin({'NotAnswered', 'NotShown', 'NotApplicable'})
       ].sort_values([f'v_{depvar}', 'lfdn'])
    var['tag'] = ''
    var.to_csv(f'../../data/freetext_tocode/v_{longvar}_with_{depvar}_total_{len(var)}.csv', 
               sep=';', index=False)
```

```python
def create_overview_df(df, longvarlist, depvarlist):
    varx = ['v_' + str(x) for x in depvarlist + longvarlist]
    overview = pd.DataFrame(columns=['lfdn','problem','reason'])
    for idx, var in enumerate(varx[0:int(len(varx)/2)]):
        overview = overview.append(
            df[['lfdn'] + [var, varx[idx+5]]].rename({var:'problem',varx[idx+5]:'reason'}, axis=1))
    return overview[~overview.reason.isin({'NotShown','NotAnswered','NotApplicable'})
            ].sort_values(['problem','lfdn'])
```
