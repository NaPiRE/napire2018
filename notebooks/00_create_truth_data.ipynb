### Note
mergeddf still has magic numbers and is somewhat weird to handle.
Let's make it nice and lean and unmagic.

-66: NotShown

-99: NotAnswered

Plus, only interesting columns, please.

```python
def kill_the_magic(mergeddf):
    newdf = mergeddf.replace({-66:'NotShown','-66':'NotShown',-99:'NotAnswered','-99':'NotAnswered'})
    return newdf

def integrate_translated_columns(mergeddf, varnames):
    newdf = pd.DataFrame(mergeddf, copy=True)
    for var in varnames:
        newdf[f'v_{var}'] = [mergeddf[f'v_{var}_translation'].values[x] 
             if mergeddf[f'v_{var}_translation'].values[x] != 'NotApplicable' else mergeddf[f'v_{var}'].values[x]
             for x in range(len(mergeddf))]
    return newdf
```

### This is not an ideal sort since it 'messes up' some of the disorganized 'order' in the survey variables...

```python
mergeddf = kill_the_magic(mergeddf)
mergeddf.head()
```
